```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
train=pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/forest-fires/forestfires.csv")
```

```python
train.head(100000)
```

```python
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()
train['month']=encoder.fit_transform(train['month'])
train['day']=encoder.fit_transform(train['day'])
```

```python
train.isnull().sum()
```

```python
features=train.drop("area",axis=1)
outcomes=train["area"].values
features.head(5)
```

```python
for i in features.columns: 
    scaler = preprocessing.StandardScaler()
    features[i] = scaler.fit_transform(features[i])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(features, outcomes, test_size=0.25, random_state=10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
from sklearn import metrics
from matplotlib import pylab

final_model = RandomForestRegressor(n_estimators=200,max_features=None,bootstrap=True,oob_score=True,max_depth=8)
final_model.fit(X_train.values, Y_train)
y_pred = final_model.predict(X_test.values)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import explained_variance_score

print(mean_squared_error(Y_test, y_pred))
print(explained_variance_score(Y_test, y_pred))
```

```python
from keras.models import Sequential
from keras.callbacks import EarlyStopping, ModelCheckpoint,History
from keras.layers import Dense, Activation, Dropout, Input
from keras import optimizers
import keras
from keras.models import Model
```

```python
def new_model():
    input_1 = Input(shape=(1,))
    input_2 = Input(shape=(1,))
    input_3 = Input(shape=(1,))
    input_4 = Input(shape=(1,))
    input_5 = Input(shape=(1,))
    input_6 = Input(shape=(1,))
    input_7 = Input(shape=(1,))
    input_8 = Input(shape=(1,))
    input_9 = Input(shape=(1,))
    input_10 = Input(shape=(1,))
    input_11= Input(shape=(1,))
    input_12= Input(shape=(1,))
    
    
    hidden_1 = Dense(32, activation='sigmoid')(input_1)
    hidden_2 = Dense(32, activation='sigmoid')(input_2)
    hidden_3 = Dense(32, activation='sigmoid')(input_3)
    hidden_4 = Dense(32, activation='sigmoid')(input_4)
    hidden_5 = Dense(32, activation='sigmoid')(input_5)
    hidden_6 = Dense(32, activation='sigmoid')(input_6)
    hidden_7 = Dense(32, activation='sigmoid')(input_7)
    hidden_8 = Dense(32, activation='sigmoid')(input_8)
    hidden_9 = Dense(32, activation='sigmoid')(input_9)
    hidden_10 = Dense(32, activation='sigmoid')(input_10)
    hidden_11 = Dense(32, activation='sigmoid')(input_11)
    hidden_12 = Dense(32, activation='sigmoid')(input_12)
    
    
    
    x = keras.layers.concatenate([hidden_1,hidden_2,hidden_3,hidden_4,hidden_5,
                                 hidden_6,hidden_7,hidden_8,hidden_9,hidden_10,
                                 hidden_11,hidden_12])
    x = Dense(256, activation='relu')(x)
    x = Dense(1)(x)
    model = Model(inputs=[input_1,input_2,input_3,input_4,input_5,
                          input_6,input_7,input_8,input_9,input_10, input_11,input_12], outputs=[x])
    model.compile(loss='mean_squared_error', optimizer = 'adam')
    return model

value_list=[X_train[['X']].values,
            X_train[['Y']].values,
            X_train[['month']].values,
            X_train[['day']].values,
            X_train[['FFMC']].values,
            X_train[['DMC']].values,
            X_train[['DC']].values,
            X_train[['ISI']].values,
            X_train[['temp']].values,
            X_train[['RH']].values,
            X_train[['wind']].values,
            X_train[['rain']].values,
           ]
    
value_list_test=[X_test[['X']].values,
            X_test[['Y']].values,
            X_test[['month']].values,
            X_test[['day']].values,
            X_test[['FFMC']].values,
            X_test[['DMC']].values,
            X_test[['DC']].values,
            X_test[['ISI']].values,
            X_test[['temp']].values,
            X_test[['RH']].values,
            X_test[['wind']].values,
            X_test[['rain']].values,
           ]
```

```python
estimator = KerasRegressor(build_fn = new_model,epochs=200,batch_size=32,)
estimator.fit(value_list, Y_train )
y_pred = estimator.predict(value_list_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import explained_variance_score

print(np.sqrt(mean_squared_error(Y_test, y_pred)))
print(explained_variance_score(Y_test, y_pred))
```
